```python
import pandas as pd
import holidays
import numpy as np

# =========================
# LOAD ORIGINAL TRAFFIC DATA
# =========================

df = pd.read_csv("Dataset_Uber Traffic.csv")

# Convert DateTime column
df['DateTime'] = pd.to_datetime(df['DateTime'])

# =========================
# TIME FEATURES
# =========================

# Hour
df['Hour'] = df['DateTime'].dt.hour

# Day Name
df['Day'] = df['DateTime'].dt.day_name()

# Month
df['Month'] = df['DateTime'].dt.month

# Weekend Feature
df['Is_Weekend'] = np.where(
    df['DateTime'].dt.dayofweek >= 5,
    1,
    0
)

# Peak Hour Feature
df['Peak_Hour'] = np.where(
    (
        ((df['Hour'] >= 7) & (df['Hour'] <= 10)) |
        ((df['Hour'] >= 17) & (df['Hour'] <= 20))
    ),
    1,
    0
)

# =========================
# WEATHER FEATURES
# =========================

np.random.seed(42)

# Temperature
df['Temperature'] = np.random.randint(
    20,
    35,
    size=len(df)
)

# Rainfall
df['Rainfall'] = np.random.choice(
    [0, 0, 0, 0.2, 0.5, 1.0, 2.0],
    size=len(df)
)

# Humidity
df['Humidity'] = np.random.randint(
    50,
    95,
    size=len(df)
)

# Wind Speed
df['WindSpeed'] = np.random.randint(
    2,
    25,
    size=len(df)
)

# =========================
# EVENT FEATURES
# =========================

# India Holidays
india_holidays = holidays.India()

# Holiday Feature
df['Is_Holiday'] = df['DateTime'].dt.date.apply(
    lambda x: 1 if x in india_holidays else 0
)

# Simulated Special Events
df['Special_Event'] = np.random.choice(
    [0, 0, 0, 0, 1],
    size=len(df)
)

# Event Impact
df['Event_Impact'] = np.where(
    (df['Special_Event'] == 1) &
    (df['Peak_Hour'] == 1),
    "High",
    "Normal"
)

# =========================
# SAVE FINAL DATASET
# =========================

df.to_csv(
    "final_traffic_event_dataset(1).csv",
    index=False
)

# =========================
# OUTPUT
# =========================

print(df.head())

print("\nHoliday Counts:")
print(df['Is_Holiday'].value_counts())

print("\nWeekend Counts:")
print(df['Is_Weekend'].value_counts())

print("\nDataset Shape:")
print(df.shape)

print("\nFinal dataset saved successfully!")
```

```python
df = pd.read_csv("Integrated_Uber_Traffic_Dataset.csv")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.fillna(0, inplace=True)
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

columns_to_scale = [
    'Vehicles',
    'Temperature',
    'Rainfall',
    'Humidity',
    'WindSpeed'
]

df[columns_to_scale] = scaler.fit_transform(
    df[columns_to_scale]
)
```
